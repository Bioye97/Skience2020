## Prelimiaries: linear inverse problems

Imagine you want to solve a linear problem with finite data $\bf d$ and finite model parameters $\bf m$. The forward operator $\bf G$, sometimes called theory matrix, is of dimension (NxM). Additionally there may be some errors $\bf e$

\begin{equation}
\bf d= Gm + e 
\end{equation}

The most general solution $\bf \tilde m$ to this problem is a linear combination of the data, expressed by an inverse operator $\bf L$ such that

\begin{equation}
\bf \tilde m = Ld
\end{equation}

If you start the solution from an intitial guess $\bf m_0$, you'll find 

\begin{equation}
\bf \tilde m = m_0 + L (d-Gm_0)
\end{equation}

or by plugging the forward equation back in

\begin{equation}
\bf \tilde m = m + (R-I) (m-m_0) + L e
\end{equation}

where $\bf R = LG$ is the resolution operator. This equation tells us that the solution you estimate $\bf \tilde m$ deviates from the solution you want $\bf m$ by 2 terms: 

**1. the imperfect resolution**

**2. the progagtion of data errors or forward modelling errors**




## Standard linear inverse operators

- The least-squares (LS) operator

\begin{equation}
\bf L=(G^tG)^{-1}G^t
\end{equation}

- The minimum norm (MN) operator

\begin{equation}
\bf L=G^t(GG^t)^{-1}
\end{equation}

- The damped least-squares (DLS) operator

\begin{equation}
\bf L=(G^tG+\theta I)^{-1}G^t = G^t(GG^t +\theta I)^{-1}
\end{equation}


## A simple example

\begin{equation}
m_1 = 1 \\
m_2 = 2 \\
m_1 + m_2 =2
\end{equation}

Let's find a solution using a least-squares inverse operator, assuming it exists.

```python
import numpy as np # Numpy Library
from numpy.linalg import inv # Numpy inverse matrix operator
import matplotlib.pyplot as plt # Plotting library

# Set some parameters to make plots nicer.
plt.rcParams["font.family"] = "serif"
plt.rcParams["figure.figsize"] = [9.6, 7.2]
plt.rcParams.update({'font.size': 20})
```

```python
G = np.array([[1, 0], [0, 1], [1, 1]])
d = np.array([[1],[2],[2]])
Gt = G.transpose()
GtG = np.matmul( Gt, G)
GtGinv = inv(GtG)
L = np.matmul(GtGinv, Gt)
mest = np.matmul(L, d)
print(f'{mest}')
```

## Geometrical interpretation of the solution

```python
#fig = plt.figure()
#ax = plt.axes()
x = np.linspace(-2, 4, 10)
plt.plot(x-x+1, x, '-k', label ='eq.1')  # eq. 1
plt.plot(x, x-x+2, '--k', label = 'eq.2')  # eq. 2
plt.plot(x, 2 -x, '-.k', label='eq.3')  # eq. 3
plt.plot(mest[0],mest[1],'r*', label='solution') # mest
plt.xlim(-1,2)
plt.ylim(0,3)
plt.xlabel('m_1')
plt.ylabel('m_2')
plt.title('Fig. 1')
plt.legend()
plt.show()
```

## Scaled equations

Let's multiply the last equation by a factor of 2, giving a new system

\begin{equation}
m'_1 = 1 \\
m'_2 = 2 \\
2 m'_1 + 2 m'_2 =4
\end{equation}

This system is mathematically equivalent and has the has same geometrical interpretation as on Fig.1

Let's again find the least-squares solution and check if $\bf m$ and $\bf m'$ are equivalent.

```python
Gprime = np.array([[1, 0], [0, 1], [2, 2]])
dprime = np.array([[1],[2],[4]])
Gtprime = Gprime.transpose()
GtGprime = np.matmul( Gtprime, Gprime)
GtGinvprime = inv(GtGprime)
Lprime = np.matmul(GtGinvprime, Gtprime)
mestprime = np.matmul(Lprime, dprime)
print(f'{mestprime}')
```

## Geometrical interpretation of the new solution

```python
#fig = plt.figure()
#ax = plt.axes()
x = np.linspace(-2, 4, 10)
plt.plot(x-x+1, x, '-k', label ='eq.1')  # e.q 1
plt.plot(x, x-x+2, '--k', label = 'eq.2')  # e.q 2
plt.plot(x, 2 -x, '-.k', label='eq.3')  # eq. 3
plt.plot(mest[0],mest[1],'r*', label='old solution') # mest
plt.plot(mestprime[0],mestprime[1],'b*', label='new solution') # mest
plt.xlim(-1,2)
plt.ylim(0,3)
plt.xlabel('m_1')
plt.ylabel('m_2')
plt.title('Fig. 2')
plt.legend()
plt.show()
```

## Can damping make the problem consistent?

Let's use 

\begin{equation}
\bf L=(G^tG+\theta I)^{-1}G^t
\end{equation}

and loop over $\theta$ to see whether $\bf m'$ now converges towards $\bf m$.

```python
#fig = plt.figure()
#ax = plt.axes()
x = np.linspace(-2, 4, 10)
plt.plot(x-x+1, x, '-k', label ='eq.1')  # e.q 1
plt.plot(x, x-x+2, '--k', label = 'eq.2')  # e.q 2
plt.plot(x, 2 -x, '-.k', label='eq.3')  # eq. 3
plt.plot(mest[0],mest[1],'r*', label='old solution') # mest
for theta in range(0,10):
    A = np.array([[theta, 0], [0, theta]])
    GtGnew = GtGprime + A
    GtGinvnew = inv(GtGnew)
    Lnew=np.matmul(GtGinvnew, Gtprime)
    mestnew = np.matmul(Lnew, dprime) 
    plt.plot(mestnew[0],mestnew[1],'b*') # mest
    plt.text(mestnew[0]+0.03,mestnew[1]-0.04,str(theta), fontsize=10)
    theta+=0.1
plt.plot(mestnew[0],mestnew[1],'b*', label='damped solutions') # mest
plt.xlim(-1,2)
plt.ylim(0,3)
plt.xlabel('m_1')
plt.ylabel('m_2')
plt.title('Fig. 3')
plt.legend()
plt.show()
```

## Regularization is still the way to make the probelm consistent

Assume the usual linear problem $\bf d=Gm$ with a damped least-squares solution $\bf \tilde m = (G^tG+\theta I)^{-1}G^t d$, which corresponds to the minimization of the cost function $C = \bf (d-Gm)^t(d-Gm) + \theta m^tm$

Now let's consider the linear transformations:

\begin{equation}
\bf d'=Qd \\
m'= Sm
\end{equation}

Both transformation matrices $\bf Q$ and $\bf S$ are invertible. This leads to a new system $\bf d'=G'm'$ where $\bf G'=QGS^{-1}$. This system of equations has the dampled least-squares solution $\bf \tilde m' = (G'^tG'+\theta I)^{-1}G'^t d'$.

Transfroming back to the original parametrization gives:

\begin{equation}
\bf \tilde m = (G^tQ^tQG+\theta S^tS)^{-1}G^t (Q^tQ)d
\end{equation}

which corresponds to the minimization of the cost function

\begin{equation}
C = \bf (d-Gm)^tQ^tQ(d-Gm) + \theta m^tS^tSm
\end{equation}

We see that $\bf Q^tQ$ and $\bf S^tS$ act as metrics for the Euclidian norms in the data and model space, respectively.

This analysis provides the reason for the inconsistency. LS assumes a unit metric when measuring data misfit and model size, regardless of the parametrization.



## Making the problem consistent

There is a need for an explicit specification of the norms leading to a more general regularization. I suggest to always mininmize the cost function:

\begin{equation}
C = \bf (d-Gm)^tW_d(d-Gm) + m^tW_mm
\end{equation}

Upon linear reparametrization $\bf d'=Qd$ and $\bf m'= Sm$, we need to change the data and model weighting to $\bf W'_d=(Q^t)^{-1} W_d Q^{-1}$ and $\bf W'_m=(S^t)^{-1} W_m S^{-1}$. This guarantees the cost function to be parametrization independent, making the solutions consistent.

In the example above $\bf S=I$ and $\bf Q =\begin{pmatrix} 1 & 0 & 0\\0 & 1& 0\\0 & 0& 2\end{pmatrix}$

```python
Q = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 2]])
Gprime = np.matmul( Q, G)
dprime = np.matmul( Q, d)
Gtprime = Gprime.transpose()
Qinv=inv(Q)
Qt = Q.transpose()
Qtinv=inv(Qt)
Wdprime=np.matmul( Qtinv, Qinv) # Wd = I, Wm = 0
GtWdprime = np.matmul( Gtprime, Wdprime)
GtWdGprime = np.matmul( GtWdprime, Gprime)
GtWdGinvprime = inv(GtWdGprime)
Lprime = np.matmul(GtWdGinvprime, GtWdprime)
mestagain = np.matmul(Lprime, dprime) # no transforming back as S=I
print(f'{mestagain}')
```

```python
#fig = plt.figure()
#ax = plt.axes()
x = np.linspace(-2, 4, 10)
plt.plot(x-x+1, x, '-k', label ='eq.1')  # e.q 1
plt.plot(x, x-x+2, '--k', label = 'eq.2')  # e.q 2
plt.plot(x, 2 -x, '-.k', label='eq.3')  # eq. 3
plt.plot(mest[0],mest[1],'ro', label='old solution') # mest
plt.plot(mestprime[0],mestprime[1],'b*', label='inconsisten solution') # mest
plt.plot(mestagain[0],mestagain[1],'g*', label='new consisten solution') # mest
plt.xlim(-1,2)
plt.ylim(0,3)
plt.xlabel('m_1')
plt.ylabel('m_2')
plt.title('Fig. 4')
plt.legend()
plt.show()
```

## The explicit Tarantolian formulation

The solution is set up as a Bayesain inference problem assuming explict Gaussian priors, leading to the cost function

\begin{equation}
C = \bf (d-Gm)^tC_d^{-1}(d-Gm) + m^tC_m^{-1}m
\end{equation}

The solution is 

\begin{equation}
\bf \tilde m = (G^tC_d^{-1}G+C_m^{-1})^{-1}G^t C_d^{-1}d
\end{equation}

Upon linear reparametrization $\bf d'=Qd$ and $\bf m'= Sm$, the above analysis tells us that we need to change the data and model covariance to $\bf C'_d=QC_d Q^t$ and $\bf C'_m=S C_m S^t$, which corresponds to the well known transformation rule of covariance matrices known from statistics.
